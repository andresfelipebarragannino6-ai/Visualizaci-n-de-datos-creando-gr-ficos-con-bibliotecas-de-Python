##**Desafío: personalizando los subplots con los datos de ventas de tiendas diferentes**

```python
import matplotlib.pyplot as plt

# Datos
tiendas = ['A', 'B', 'C', 'D']
meses = list(ventas_2022.keys())
ventas = list(ventas_2022.values())

# Colores personalizados por tienda
colores = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']

# Crear figura y subgráficos
fig, axs = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle(
    'Variación de Ventas Mensuales por Tienda - 2022',
    fontsize=20,
    fontweight='bold'
)

# Bucle para recorrer los subgráficos
for i, ax in enumerate(axs.flat):
    ventas_tienda = [mes[i] for mes in ventas]

    ax.plot(
        meses,
        ventas_tienda,
        color=colores[i],
        linewidth=3
    )

    # Títulos y etiquetas
    ax.set_title(
        f'Tienda {tiendas[i]}',
        loc='left',
        fontsize=14,
        fontweight='bold'
    )
    ax.set_xlabel('Mes', fontsize=12)
    ax.set_ylabel('Ventas', fontsize=12)

    # Personalización adicional
    ax.tick_params(axis='both', labelsize=10)
    ax.grid(True, linestyle='--', alpha=0.5)

# Ajustar espacios
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

##**Desafío: creando un gráfico de líneas con Seaborn**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(
    style="whitegrid",
    context="talk",
    font_scale=1.1
)

sns.set_palette("colorblind")

df = pd.read_csv('/content/inmigrantes_canada.csv')

anos_columnas = [str(year) for year in range(1980, 2014)]


paises = ["Brasil", "Argentina", "Perú", "Colombia"]
df_latam_filtered = df[df["Pais"].isin(paises)].copy()

df_melted = df_latam_filtered.melt(
    id_vars=['Pais', 'Continente', 'Region'],
    value_vars=anos_columnas,
    var_name='Año',
    value_name='Inmigrantes'
)

df_melted['Año'] = pd.to_numeric(df_melted['Año'])

plt.figure(figsize=(12, 6))

sns.lineplot(
    data=df_melted,
    x="Año",
    y="Inmigrantes",
    hue="Pais",
    linewidth=2.5
)

plt.title(
    "Tendencias de Inmigración desde América Latina",
    fontsize=18,
    pad=15
)

plt.xlabel("Año", fontsize=14)
plt.ylabel("Número de inmigrantes", fontsize=14)

plt.legend(
    title="País",
    title_fontsize=13,
    fontsize=12,
    loc="upper left"
)

plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objs as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=[datos['Año'].iloc[0]],
    y=[datos['Argentina'].iloc[0]],
    mode='lines+markers',
    name='Argentina',
    line=dict(width=3)
))

fig.add_trace(go.Scatter(
    x=[datos['Año'].iloc[0]],
    y=[datos['Brasil'].iloc[0]],
    mode='lines+markers',
    name='Brasil',
    line=dict(width=3)
))
fig.update_layout(
    title=dict(
        text='Inmigración de Argentina y Brasil hacia Canadá (1980–2013)',
        x=0.5,              # 🎯 centrado horizontal
        xanchor='center'
    ),
    xaxis=dict(title='Año', range=[1980, 2013]),
    yaxis=dict(title='Número de Inmigrantes'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        x=-0.05,
        y=1.18,
        xanchor='left',
        yanchor='top',
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {
                'frame': {'duration': 120, 'redraw': True},
                'fromcurrent': True
            }]
        )]
    )],
    width=900,
    height=450
)

frames = []

for i in range(len(datos)):
    frames.append(
        go.Frame(
            data=[
                go.Scatter(
                    x=datos['Año'].iloc[:i+1],
                    y=datos['Argentina'].iloc[:i+1]
                ),
                go.Scatter(
                    x=datos['Año'].iloc[:i+1],
                    y=datos['Brasil'].iloc[:i+1]
                )
            ]
        )
    )

fig.frames = frames
fig.show()
```
